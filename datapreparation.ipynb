```python
import pandas as pd
from googletrans import Translator

# Load the dataset
file_path = 'dataset.csv'  # Replace with your file path
dataset = pd.read_csv(file_path)

# Initialize the translator
translator = Translator()

# Translate columns to Tamil
print("Translating to Tamil... This may take time.")
dataset['question'] = dataset['question'].apply(lambda x: translator.translate(x, src='en', dest='ta').text)
dataset['answer'] = dataset['answers'].apply(lambda x: translator.translate(x, src='en', dest='ta').text)

# Save only the Tamil columns to a new CSV file
translated_file_path = 'dataset_tamil.csv'
dataset[['question', 'answer']].to_csv(translated_file_path, index=False, encoding='utf-8')

print(f"Translation complete! File saved as '{translated_file_path}'")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load your dataset with the correct encoding
file_path = 'dataset_tamil_agri.csv'  # Replace with your dataset path
dataset = pd.read_csv(file_path, encoding='ISO-8859-1')  # Use detected encoding

# Split the dataset into train and temp (temporary for validation and test)
train, temp = train_test_split(dataset, test_size=0.3, random_state=42)

# Split temp into validation and test
validation, test = train_test_split(temp, test_size=0.5, random_state=42)

# Save each split to separate CSV files
train.to_csv('train.csv', index=False, encoding='utf-8')
validation.to_csv('validation.csv', index=False, encoding='utf-8')
test.to_csv('test.csv', index=False, encoding='utf-8')

print("Dataset split into train.csv, validation.csv, and test.csv!")
```
